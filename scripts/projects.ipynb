```python
import pandas as pd
import numpy as np
```

```python
# read from excel

df = pd.read_excel('卡牌数据&数学模型.xlsx', sheet_name='保护项目牌')
```

```python
df = df[df['编号'].notna()] # 过滤无效信息

# float to 3-char string, such as 1.0 -> '001'
df['编号'] = df['编号'].apply(lambda x: str(int(x)).zfill(3))

df.head()
```

```python
# 定义生成文本内容的函数
def generate_text(row):
    new_string = f"""
  {{
    id: '{row['编号']}',
    name: "{row['英文名']}",
    cnName: "{row['中文名']}",
    image: '/img/projects/{row['编号']}{row['中文名']}.png',
    type: ProjectCategory.BASE,
    source: CardSource.BASE,
  }},
"""
    return new_string

with open('projects.txt', 'w') as f:
    for index, row in df.iterrows():
        new_string = generate_text(row)
        f.write(new_string)
```
